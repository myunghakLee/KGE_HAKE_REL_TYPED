```python
from glob import glob


files = sorted(glob("models/*/*/train.log"))

```

```python
from matplotlib import pyplot as plt

def draw_graph(files):
    for file in files:
        if "OVERFITTING" in file:
            continue
        if "ModE" in file:
            continue
        f = open(file)
        data = f.readlines()
        hits0 = []
        for line in data:
            if "HITS@1 " in line:
                hits0.append(float(line[-8:]))
        if "ModE" in file:
            plt.plot(hits0, label = file.split("/")[2], color= 'red', linewidth=1)
        else:
            plt.plot(hits0, label = file.split("/")[2], color= 'blue', linewidth=1)
        f.close()
    plt.legend(bbox_to_anchor=(1.1, 1.05))
    plt.show()
```

```python
draw_graph(files)
```

```python
from matplotlib import pyplot as plt

def draw_graph(files):
    for file in files:
        if "OVERFITTING" in file:
            continue
        if "OLD" in file:
            continue
        if "ModE" in file:
            continue
#         if "(h*" in file:
#             continue
        f = open(file)
        data = f.readlines()
        hits0 = []
        for line in data:
            if "HITS@1 " in line:
                hits0.append(float(line[-8:]))
        hits0 = hits0[4:]
        if "Multi" in file and "(h*" in file:
            plt.plot(hits0, label = file.split("/")[1]+file.split("/")[2], color= '#FF9999', linewidth=1)
        elif "Multi" in file:
            plt.plot(hits0, label = file.split("/")[1]+file.split("/")[2], color= '#FF0000', linewidth=1)
        elif "Multi" not in file and "(h*" in file:
            plt.plot(hits0, label = file.split("/")[1]+file.split("/")[2], color= '#9999FF', linewidth=1)
        else:
            plt.plot(hits0, label = file.split("/")[1]+file.split("/")[2], color= 'blue', linewidth=1)
        f.close()
    plt.legend(bbox_to_anchor=(1.1, 1.05))
    plt.show()
```

```python
draw_graph(files)
```

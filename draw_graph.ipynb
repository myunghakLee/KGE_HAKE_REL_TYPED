```python
from glob import glob


files = sorted(glob("models/ModE_DocRED_triple*/train.log"))

```

```python
from matplotlib import pyplot as plt

def draw_graph(files):
    for file in files:
        f = open(file)
        data = f.readlines()
        hits0 = []
        for line in data:
            if "HITS@1 " in line:
                hits0.append(float(line[-8:]))
        if "Not" in file:
            plt.plot(hits0, label = file.split("/")[1], color= 'red')
        else:
            plt.plot(hits0, label = file.split("/")[1], color= 'blue')
        f.close()
    plt.legend(bbox_to_anchor=(1.1, 1.05))
    plt.show()
```

```python
draw_graph(files)
```
